### Sitmulation can be useful to understan the important factors( perform sensitivity analysis)

```python
## basic simtulation 

totalPopulation = 50 
growthFactor=1.000005
dayCount=0 #every 2 month population is reported

while totalPopulation<1000000:
    totalPopulation*=growthFactor   # totalPopulation=totalPopulation*growthFactor
    
    #every 56th day population is reported
    
    dayCount +=1   # dayCount=dayCount+1
    if dayCount==56:
        dayCount=0
        print(totalPopulation)
```
